2. Find the Runner-Up Score!

```python
arr = [7,6, 6, 6, 6, 6, 6, 6, 6, 6, 5]

print(len(arr))
sor_array = sorted(arr)
max = sor_array[-1]

runner = 0
for i in sor_array:
  if i < max:
    runner = i

print("Final = %d"%(runner))

```

3. Nested Lists

```python
records = [["Harry",37.21],["Berry",37.22],["Tina",37.20],["Akriti",41],["Harsh",39]]

max = 0.0
sec_max = 0.0
thr_max = 0.0

for i in records:
   print(float(i[1]))    
   if (float(i[1]) > max):
     max = float(i[1])
   elif (sec_max < float(i[1])):
     sec_max = float(i[1])
   elif (thr_max < float(i[1])):
     thr_max = float(i[1])


print("%.2f" % max)
print("%.2f" % sec_max)
print("%.2f" % thr_max)

lis = []
for i in records:
  if(i[1] == thr_max):
    lis.append(i[0])

print(lis)
for i in records:
  print(i)
```

Finding the percentage

```python
n = int(input())
student_marks = {}
for _ in range(n):
  line = input().split()
  name, scores = line[0], line[1:]
  scores = map(float, scores)
  student_marks[name] = scores
query_name = input()
sum = 0

for val in student_marks[query_name]:
  sum += val

print('%.2f'%(sum/3))
```

Lists

```python
n = int(input())
lists = []

for _ in range(n):
  line = input().split()
  oper, values = line[0], line[1:]
  
  if(oper == 'insert'):
    lists.insert(int(values[0]),int(values[1]))

  if(oper == 'print'):
    print(lists) 

  if(oper == 'remove'):
    lists.remove(int(values[0]))

  if(oper == 'append'):
    lists.append(int(values[0]))

  if(oper == 'sort'):
    lists.sort()

  if(oper == 'pop'):
    lists.pop()  

  if(oper == 'reverse'):
    lists.reverse()      
```
