sWAP cASE



```python
def swap_case(s):
    strings = ''
    for i in range(len(s)):
        if s[i].isupper():
            strings =  strings + s[i].lower()
        else:
            strings =  strings + s[i].upper()
                
    return strings
 
print(swap_case("Hello")) 
```

String Split and Join

```python
def split_and_join(line):
    line = line.split(" ")
    return "-".join(line)

print(split_and_join("this is a string"))   
```

```python
def mutate_string(string, position, character):
    return string[:position] + (character) + string[position+1:]

print(mutate_string("abracadabra",5,"k"))    
```

Find a string

```python
def count_substring(string, sub_string):
    count = 0
    
    for i in range(len(string)):
        if(string[i:i+len(sub_string)] == sub_string):
            count += 1
            
    return count

print(count_substring("ABCDCDC","CDC")) 
```

String Validators

```python
s = input()
    
l = [False,False,False,False,False]
for i in range(len(s)):
  if  (s[i].isalnum()):
    l[0] = True
  if (s[i].isalpha()):
    l[1] = True
  if (s[i].isdigit()):
    l[2] = True
  if (s[i].islower()):
    l[3] = True
  if (s[i].isupper()):
    l[4] = True
    
for o in l:
  print(o)
```

Text Wrap

```python
def wrap(string , max_width):
    li = ''
    i = 0
    while( i < len(string)):
        if ((i + max_width) < len(string)):
            li += (string[i:i+max_width] + '\n')
        else:    
            li += (string[i:i+max_width])
        i += max_width

    li += (string[i:-1])

    return li 

print(wrap("ABCDEFGHIJKLIMNOQRSTUVWXYZ", 4)) 
```
